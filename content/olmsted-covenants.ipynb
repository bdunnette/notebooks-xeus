```python
# Ensure that the necessary packages are installed
# Load Geopandas
import geopandas as gpd
from folium.plugins import Fullscreen, MousePosition
```

```python
try:
    import pyodide_http

    pyodide_http.patch_all()
except ImportError:
    pass  # If patching fails, we can still proceed without it
```

```python
# Load covenants into a GeoPandas dataframe
olmsted_covenants = gpd.read_file('data/covenants-mn-olmsted-county.zip')
olmsted_covenants.head()
```

```python
# Map the covenants on a Folium map
map = olmsted_covenants.explore()
Fullscreen(
    position="topright",
    title="Expand me",
    title_cancel="Exit me",
    force_separate_button=True,
).add_to(map)
MousePosition().add_to(map)

map
```
